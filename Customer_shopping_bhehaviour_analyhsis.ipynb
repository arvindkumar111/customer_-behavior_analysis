```python
# fill the null value in the review rating with median in same category
data['Review Rating'] = data.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
data.isnull().sum()
```

```python
# the column name is in wrong format change to snake case
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(' ','_')
```

```python
data.columns
```

```python
# create new column age_group
labels = ['Young_Adult' , 'Adult', 'Middle_aged' , 'Senior']
data['age_group'] = pd.qcut(data['age'],q = 4 , labels = labels)
```

```python
# Now Create the new colum purchase_frequency_days
## We are making this column to change text of purchase_frequency to numeric 
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months'  : 90
}
    
data['purchase_frequency_column'] = data['frequency_of_purchases'].map(frequency_mapping)
```

```python
data[['purchase_frequency_column','frequency_of_purchases']]
```

```python
# Now find out "discount_applied" and "promo_code_used" is same or not
(data['discount_applied'] == data['promo_code_used']).all()
```

```python
# yes above both columns have same value so drop 1 
data=data.drop('promo_code_used',axis=1)
```
